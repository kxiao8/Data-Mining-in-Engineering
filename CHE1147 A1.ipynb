### 1. Generate a sample of 10 random integer numbers between 600 and 900 (hint: use Numpy’s random integers function).

### 2. Create a function called credit score that reads the sample and returns the output ’Low’ when the input is [600, 699], ’Medium’ when in [700, 799] and ’High’ when in [800, 900].

```python
# Assign categorical values within certain range
def credit_score(ls):
    out = []
    for e in ls:
        if 600 <= e <= 699:
            out.append('Low')
        elif 700 <= e <= 799:
            out.append('Medium')
        elif 800 <= e <= 900:
            out.append('High')
        else:
            return 'Value Error'
    return out

# run
credit_score(rand_array)
```

##  2.2 Norms and eigenvalues
### 1. Calculate the maximum and Euclidean or Frobenius norm of each of the three vectors:

```python
x_1 = np.array([1, -2, 3]).reshape(-1,1)
x_2 = np.array([2, 0, -1, 2]).reshape(-1,1)
x_3 = np.array([0, 1, -4, 2, -1]).reshape(-1,1)

vector_ls = [x_1, x_2, x_3]

for i in vector_ls:
    print('Max:', i.max())
    print('Norm:', np.linalg.norm(i))
```

### 2. Calculate the l 2 and l ∞ norms of the matrices:

```python
A_1 = np.array([1, -2, 4, 3]).reshape(2,2)
A_2 = np.array([1, 0, 2, 0, 1, -1, -1, 1, 1]).reshape(3,3)

vector_ls2 = [A_1, A_2]

for i in vector_ls2:
    print('L-2 Norm:', np.linalg.norm(i, 2))
    print('L-inf Norm:', np.linalg.norm(i, np.inf))
```

## 3 Pandas manipulations
### 1. Import file Gas prices.xls as a pandas DataFrame. Create a function that replaces NaN with the interpolation between the adjacent values (i.e. the prior and the after the NaN). Do not use existing interpolation functions, built your own.

```python
import pandas as pd

gas_prices = pd.read_excel('Gas_prices.xlsx')
gas_prices
```

```python
# Replace NaN with the adjacent values
def na_replace(df, col_name):
    for r in range(len(df[col_name])):
        if np.isnan(df[col_name][r]): # replace NaN
            df[col_name][r] = (df[col_name][r - 1] +
                               df[col_name][r + 1]) / 2

# run
na_replace(gas_prices, 'Price_pl')
gas_prices
```

### 2. Find the index, the date and the value of the min and max gas price.

```python
# set original values
price_min = np.inf
price_max = -np.inf
min_i = -1
max_i = -1

# loop through all value and save the value and indices
for i in range(len(gas_prices['Price_pl'])):
    if gas_prices['Price_pl'][i] > price_max:
        price_max = gas_prices['Price_pl'][i].copy()
        max_i = i
    if gas_prices['Price_pl'][i] < price_min:
        price_min = gas_prices['Price_pl'][i].copy()
        min_i = i

print('Min Index:', min_i)
print('Min Date:', gas_prices['Date'][min_i])
print('Min Price:', price_min)
print('Max Index:', max_i)
print('Max Date:', gas_prices['Date'][max_i])
print('Max Price:', price_max)
```

### 3. Calculate the following descriptive statistics: mean, median, quantile, skewness and kurtosis.

```python
print('Mean Price:', gas_prices['Price_pl'].mean())
print('Median Price:', gas_prices['Price_pl'].median())
print('Price Quantile:')
print(gas_prices.describe()['Price_pl'][4:7])
print('Price Skewness:', gas_prices['Price_pl'].skew())
print('Price Kurtosis:', gas_prices['Price_pl'].kurtosis())
```

### 4. Create two new columns with the Month and Year. Pivot the table by Month(index) and Year(columns). Calculate and plot the monthly average gas price. This should give you one line per year.

```python
# save month and year value to lists
gas_month = []
gas_year = []
for s in gas_prices['Date']:
    ls_date = s.split('/')
    gas_month.append(int(ls_date[1]))
    gas_year.append(ls_date[2])

# assign to gas_prices dataframe
gas_prices['Month'] = gas_month
gas_prices['Year'] = gas_year
```

```python
# calculate monthly average with groupby()
gas_prices_monthly_ave = gas_prices.groupby(['Year', 'Month'])['Price_pl'].mean()

# create list to save the values from series created by groupby()
gas_monthly_data = []

# keep the indices from the groupby()
for i in range(len(gas_prices_monthly_ave)):
    gas_monthly_data.append(gas_prices_monthly_ave.index[i]+(gas_prices_monthly_ave[i],))

# make the monthly average dataframe
gas_monthly = pd.DataFrame(gas_monthly_data, columns=['Year', 'Month', 'Price_Ave'])

# pivot the dataframe to Month/Year
gas_monthly_prices = gas_monthly.pivot(index='Month', columns='Year', values='Price_Ave').sort_index()
gas_monthly_prices
```

```python
# plot one line per year
from matplotlib import pyplot as plt
gas_monthly_prices.plot(figsize = (20,10))
plt.title("Gas Price Monthly Average", fontsize=24)
plt.xlabel("Month", fontsize=18)
plt.ylabel("Gas Price", fontsize=18)
plt.legend(title='Year', loc='best', fontsize=14, title_fontsize=16)
```

```python
# index the month and make the year as column value
TREB_monthly = TREB_data.pivot(index=['Month'], columns='Year', values='Sales')
TREB_monthly
```
